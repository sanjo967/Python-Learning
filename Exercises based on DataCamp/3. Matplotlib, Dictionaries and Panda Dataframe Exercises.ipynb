### Section 1 **Matplotlib Customization Challenge**

#### **Objective:**  
Create a **customized line plot** that visualizes the **world population growth** over time using Matplotlib.  

#### **Steps to Follow:**  
1. **Import Matplotlib.**  
2. **Create two lists**:  
   - `years`: **[1800, 1850, 1900, 1950, 2000, 2050, 2100]**  
   - `population`: **[1.0, 1.2, 1.6, 2.5, 6.1, 9.7, 10.5]** (Population in billions)  
3. **Plot the data as a line chart.**  
4. **Customize your chart:**  
   - Add **x-axis and y-axis labels**.  
   - Add a **title** to describe the chart.  
   - Adjust **y-axis ticks** to show values in **billions (e.g., 2B, 4B, etc.)**.

```python
#1. Importing matplotlib as plt
import matplotlib.pyplot as plt

#2. Creating lists
years = [1800, 1850, 1900, 1950, 2000, 2050, 2100]
population = [1.0, 1.2, 1.6, 2.5, 6.1, 9.7, 10.5]

#3. Plotting the data as a line chart
plt.plot(years, population, label= "poluation growth")

#4. Customising the chart

plt.xlabel("Years")
plt.ylabel("Population")
plt.title("World Population growth")
plt.yticks([2, 4, 6, 8, 10, 12],
           ["2B", "4B", "6B", "8B", "10B", "12B"])


#Bonus Challenging: Adding scatter plot
plt.scatter(years, population, color="red", label= "data points")

plt.legend()
plt.show()
```

### 6. Creating a DataFrame from a Dictionary

Create a dictionary called brics_data containing:
"Country" â†’ ["Brazil", "Russia", "India", "China", "South Africa"]

"Capital" â†’ ["BrasÃ­lia", "Moscow", "New Delhi", "Beijing", "Pretoria"]

"Area" â†’ [8.5, 17.1, 3.3, 9.6, 1.2] (in million kmÂ²)

"Population" â†’ [213, 144, 1391, 1444, 60] (in millions)

Convert this dictionary into a Pandas DataFrame named brics.

Set the index to ["BR", "RU", "IN", "CH", "ZA"].

```python
# Creating Brics Dictionary
brics_data = {"Country": ["Brazil", "Russia", "India", "China", "South Africa"],
              "Capital": ["BrasÃ­lia", "Moscow", "New Delhi", "Beijing", "Pretoria"],
              "Area": [8.5, 17.1, 3.3, 9.6, 1.2],
              "Population": [213, 144, 1391, 1444, 60],} #Setting up columns

#Converting dictionary to dataframe
import pandas as pd
brics=pd.DataFrame(brics_data)


#Setting index as ["BR", "RU", "IN", "CH", "ZA"].
brics.index=["BR", "RU", "IN", "CH", "ZA"]

#Printing brics to check
print(brics)
```
