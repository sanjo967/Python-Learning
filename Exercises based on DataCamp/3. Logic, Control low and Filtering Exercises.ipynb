## **Exercise 2: Filtering with NumPy Arrays**

1. Create a NumPy array named numbers containing the following values: [1, 5, 3, 8, 2, 7, 4, 10, 6].

2. Create a Boolean mask that selects all numbers in the array that are greater than 5.

3. Use the mask to extract only the numbers that are greater than 5 from the array.
4. Create a new mask that selects all numbers that are even.
5. Use both masks (for numbers greater than 5 and even numbers) to select the numbers that satisfy both conditions.

```python
#importing numpy
import numpy as np

# 1.Creating Numpy array
list = [1, 5, 3, 8, 2, 7, 4, 10, 6]
array=np.array(list)

#2. Create a Boolean mask
big_array=array>5

#3. Extracting numbers greater than 5
print("Then numbers greater than 5 in the array are ", array[big_array])
```

```python
#4. Only even numbers
even_array=array % 2 == 0
even_array

print(array[even_array])
```

```python
#5 Selecting numbers greater than 5 and even numbers in boolean mask
even_big_array=np.logical_and(even_array, big_array)

#Filtering array
array[even_big_array]
```

## **Exercise 3: Filtering Pandas DataFrames**

Load the "tips" dataset from the seaborn library. (You might need to install seaborn first: pip install seaborn)

1. Filter the DataFrame to select only tips that are greater than $5.

2. Filter the DataFrame to select only rows where the 'smoker' column is 'Yes'.
3. Filter the DataFrame to select rows where the 'day' is 'Sun' and the 'time' is 'Dinner'.

4. Filter the DataFrame to select rows where the 'total_bill' is greater than 20 dollars or the 'tip' is greater than 5 dollars.


```python
#Loading tips dataset
import seaborn as sns

tips_df = sns.load_dataset("tips")
tips_df
```

```python
#1. Filtering for rows with tips only greater than 5
big=tips_df["tip"]>5
big_tips=tips_df[big]
big_tips #returns back data frame with rows taht satisfy our condition
```

```python
#Filter the DataFrame to select only rows where the 'smoker' column is 'Yes'.
tips_df = sns.load_dataset("tips")
smoker=tips_df["smoker"]=="Yes"
print(smoker)
yes_smoker=tips_df[smoker]
yes_smoker
```

```python
# 4. Filter the DataFrame to select rows where the 'day' is 'Sun' and the 'time' is 'Dinner'.
#importing numpy as np
import numpy as np
sunday_dinner=np.logical_and(tips_df["day"]=="Sun", tips_df["time"]=="Dinner") #creating boolean mask
#Subsetting dataframe
sunday_dinner_tips_df=tips_df[sunday_dinner]
sunday_dinner_tips_df
```

```python
#5 Filter the DataFrame to select rows where the 'total_bill' is greater than $20 or the 'tip' is greater than $5.

#Creating boolean mask
big_bill_or_tips=np.logical_or(tips_df['total_bill']>20, tips_df['tip']>5)

#Subsetting dataframe
big_bill_or_tip_tips_df=tips_df[big_bill_or_tips]
big_bill_or_tip_tips_df
```

# **Exercise 4: Conditional Statements**

### 1. Write a Python function that takes an integer as input.

If the integer is even, the function should return "Even".

If the integer is odd, the function should return "Odd".




```python
#.1 Function to determine odd or even numbers
def even_or_odd(x):
  if x %2 == 0:
    print("Even")
  else:
    print("Odd")


#Testing function
even_or_odd(2)
```

### 2. Write a Python function that takes a number as input.

If the number is positive, return "Positive".

If the number is negative, return "Negative".

If the number is zero, return "Zero".


```python
#2 Function to determine positive or negative number
def postive_or_negative(x):
  if x > 0:
    return "Positive"
  elif x <0:
    return "Negative"
  elif x==0:
    return "Zero"

#Sanity check
postive_or_negative(0)
```
