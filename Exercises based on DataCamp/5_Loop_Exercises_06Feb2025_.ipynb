## Exercise 1: While Loop

Write a while loop that prints the numbers from 1 to 10.



```python
#Write a while loop that prints the numbers from 1 to 10.
x=1
while x <= 10:
  print(x)
  x=x + 1
```

Create a while loop that calculates the sum of all numbers from 1 to 100.

```python
#Create a while loop that calculates the sum of all numbers from 1 to 100.
x=0
list=[]
while x <= 100:
  list.append(x)
  x = x + 1

sum(list)
```

# Exercise 2: For Loop

Create a list of your favorite fruits.

Use a for loop to print each fruit in the list.

Create a string. Use a for loop to print each character of the string in uppercase.

Given a list of numbers, use a for loop to find the sum of all the numbers.

```python
# Creating list
fruits= ["apple", "pear", "orange", "banana", "kiwi"]

for fruit in fruits:
  print(fruit)
```

```python
# creating string
s="superman"
#looping letters in string
for letters in s:
  print(letters.capitalize())
```

```python
#creating list
list=[1,2,3,4,5,12,14]
total_sum = 0  # Initialize a variable to store the sum

for i in range(len(list) - 1):
  total_sum = total_sum + list[i] + list[i + 1]

print(total_sum) # Print the total sum
```

## Exercise 3: Looping with Dictionaries

Create a dictionary where the keys are names and the values are ages.

Use a for loop to print the name and age of each person in the dictionary.

Modify the loop to print only the names.

```python
# Creating dictionary
ages = { "Kevin": 24, "Tom": 56, "Sam":26, "Joe":30, "Sal":67}

#print names using loop
for name, age in ages.items():
  print( name + " is " + str(age) + " years old")
```

```python
#Modifying loop to print only names
for names in ages.items():
   print(names[0])
```

# Exercise 4: Looping with NumPy Arrays

Create a NumPy array of 10 random numbers.

Use a for loop to print each number in the array.

Create a 2D NumPy array. Use a for loop (and np.nditer()) to print each element in the array.

```python
#importing numpy
import numpy as np

#Creating array
list= [2,3,54,33,21,35,38,21,76,99]

#Converting to an array
array = np.array(list)
array

#Printing each number in the array with for loop
for n in np.nditer(array):
  print(n)
```

## Exercise 5: Looping with Pandas DataFrames

Load the "tips" dataset from seaborn.

Use a for loop and iterrows() to print the 'total_bill' and 'tip' for each row in the DataFrame.

Using a for loop and iterrows(), create a new column in the DataFrame that calculates the tip percentage.

```python
#Loading tips data set
import seaborn as sns

tips_df = sns.load_dataset("tips")
tips_df

#printing total bill and tips
for label, row in tips_df.iterrows():
  print(label, row["total_bill"], row["tip"])
```

```python
# create a new column in the DataFrame that calculates the tip percentage.
for label, row in tips_df.iterrows():
  tips_df["tip %"]=row["tip"]/row["total_bill"]

```

```python
# create a new column in the DataFrame that calculates the tip percentage.
#quicker do it this way
tips_df["tip %"] = (tips_df["tip"] / tips_df["total_bill"])
tips_df
```
